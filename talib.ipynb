```python
from talib import abstract
```

```python
pd.read_excel(company_index[0]+".xlsx")
```

```python
df = pd.read_excel(company_index[0]+".xlsx")
```

```python
inputs = {
    'open': df["open"],
    'high': df["high"],
    'low':  df["low"],
    'close': df["close"],
    'volume': df["volume"]
}
# uses close prices (default)
out1 = abstract.SMA(inputs, timeperiod=5)
out2 = abstract.SMA(inputs, timeperiod=10)
out3 = abstract.SMA(inputs, timeperiod=15)
out4upperband, out4middleband, out4lowerband = abstract.BBANDS(inputs, timeperiod=5, nbdevup=2, nbdevdn=2, matype=0)
out5 = abstract.DEMA(inputs, timeperiod=6)
out6slowk, out6slowd = abstract.STOCH(inputs, fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
out7 = abstract.ADX(inputs, timeperiod=7)
out8macd, out8macdsignal, out8macdhist = abstract.MACD(inputs, fastperiod=6, slowperiod=12, signalperiod=9)
```

```python
pd.concat([df, pd.DataFrame(out1, columns=["ma5"])\
          , pd.DataFrame(out2, columns=["ma10"])\
          , pd.DataFrame(out3, columns=["ma15"])\
          ,pd.DataFrame(out4upperband, columns=["upperband"])\
          ,pd.DataFrame(out4middleband, columns=["middleband"])\
          ,pd.DataFrame(out4lowerband, columns=["lowerband"])\
          ,pd.DataFrame(out5, columns=["DEMA"])\
          ,pd.DataFrame(out6slowk, columns=["slowk"])\
          ,pd.DataFrame(out6slowd, columns=["slowd"])\
          ,pd.DataFrame(out7, columns=["ADX"])\
          ,pd.DataFrame(out8macd, columns=["macd"])\
          ,pd.DataFrame(out8macdsignal, columns=["macdsignal"])\
          ,pd.DataFrame(out8macdhist, columns=["8macdhist"])], axis=1).head(50)
```

```python
for w in range(0, len(company_index)):
    df = pd.read_excel(company_index[w]+".xlsx")
    inputs = {
    'open': df["open"],
    'high': df["high"],
    'low':  df["low"],
    'close': df["close"],
    'volume': df["volume"]
    }
    # uses close prices (default)
    out1 = abstract.SMA(inputs, timeperiod=5)
    out2 = abstract.SMA(inputs, timeperiod=10)
    out3 = abstract.SMA(inputs, timeperiod=15)
    out4upperband, out4middleband, out4lowerband = abstract.BBANDS(inputs, timeperiod=5, nbdevup=2, nbdevdn=2, matype=0)
    out5 = abstract.DEMA(inputs, timeperiod=6)
    out6slowk, out6slowd = abstract.STOCH(inputs, fastk_period=5, slowk_period=3, slowk_matype=0, slowd_period=3, slowd_matype=0)
    out7 = abstract.ADX(inputs, timeperiod=7)
    out8macd, out8macdsignal, out8macdhist = abstract.MACD(inputs, fastperiod=6, slowperiod=12, signalperiod=9)
    a =  pd.DataFrame(out1, columns=["ma5"])
    b = pd.DataFrame(out2, columns=["ma10"])
    c = pd.DataFrame(out3, columns=["ma15"])
    d = pd.DataFrame(out4upperband, columns=["upperband"])
    e = pd.DataFrame(out4middleband, columns=["middleband"])
    f = pd.DataFrame(out4lowerband, columns=["lowerband"])
    g = pd.DataFrame(out5, columns=["DEMA"])
    h = pd.DataFrame(out6slowk, columns=["slowk"])
    i = pd.DataFrame(out6slowd, columns=["slowd"])
    j = pd.DataFrame(out7, columns=["ADX"])
    k = pd.DataFrame(out8macd, columns=["macd"])
    l = pd.DataFrame(out8macdsignal, columns=["macdsignal"])
    m = pd.DataFrame(out8macdhist, columns=["8macdhist"])
    pd.concat([df, a , b, c, d, e, f, g, h, i, j, k, l, m], axis=1).to_excel(company_index[w]+".xlsx", \
                                              sheet_name = company_index[w], index=False)
    if(w%100==0):
        print(w)
```

```python
train_set = pd.read_excel("strategy_results_train.xlsx")
```

```python
import copy
```

```python
dataset = copy.deepcopy(train_set)
```

```python
temp = []
for i in range(0, dataset.shape[0]):
# for i in range(0, 10):
    if(i%1000==0):
        print(i, datetime.datetime.now())
    trade_company = dataset.iloc[i][0].split("_")[0]
    trade_date    = dataset.iloc[i][0].split("_")[1]
    df = pd.read_excel(trade_company+".xlsx")
    trade_date_datetime = datetime.date(int(trade_date[:4]), int(trade_date[4:]), 1) - datetime.timedelta(days=1)
    lower_index = 0
    for index in range(0, df.shape[0]):
        if(df["date"].iloc[index]< trade_date_datetime):
            lower_index = index
        else:
            break
#     print(list(df.iloc[lower_index, :]))
    temp.append(list(df.iloc[lower_index, :]))
```
